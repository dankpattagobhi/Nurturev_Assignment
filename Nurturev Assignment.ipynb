```python
rows=data.shape[0]

new_column=[]

for i in range(rows):

        employee_count_by_month = data.loc[i,'employee_count_by_month_by_role']
        
        employee_count_by_month_dict = eval(employee_count_by_month)
        
        sales_count_by_month = {month: details.get('sales', 0) for month, details in employee_count_by_month_dict.items()}
        
        
        sales_df = pd.DataFrame(list(sales_count_by_month.items()), columns=['Month', 'Sales_Employees'])
        
        last_6_months_df = sales_df.tail(6)
        
        initial_count = last_6_months_df['Sales_Employees'].iloc[0]
        
        final_count = last_6_months_df['Sales_Employees'].iloc[-1]
        
        percentage_change = ((final_count - initial_count) / initial_count) * 100

        new_column.append(percentage_change)
        
    
data['6_months_decline']=new_column

data1 = data.sort_values(by=['6_months_decline'],ascending=True)
data1.head()


```

We will use relative decline due to the large variation in no. of employees it is not feasible to judge decline in sales team just by looking at absolute numbers
whereas when we look at realtive decline this gives an more broader view of the situation by normalizing the data and comparing every row equally.
